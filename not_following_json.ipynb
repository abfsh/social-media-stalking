```python
import json
import os

desktop_path = os.path.expanduser("~/Desktop/following.json")
with open(desktop_path, 'r') as file:
    following = json.load(file)

desktop_path = os.path.expanduser("~/Desktop/followers_1.json")
with open(desktop_path, 'r') as file:
    followers = json.load(file)

def parse_json(json_file, key):
    curr_set = set()
    if isinstance(json_file, dict):
        data = json_file.get(key, [])
    else:
        data = json_file

    for user in data:
        if 'string_list_data' in user and user['string_list_data']:
            curr_set.add(user['string_list_data'][0]['value'])
    return curr_set

following_set = parse_json(following, 'relationships_following')
followers_set = parse_json(followers, 'relationships_followers')

not_following = following_set - followers_set

for user_not_following in not_following:
    print(user_not_following)
```
